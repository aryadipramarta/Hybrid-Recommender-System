```python
import pandas as pd
import numpy as np
import seaborn as sns
import glob
import os
import random
```

```python
path = r'E:\KULIAH\SEMESTER 7\TA\dataset_user'
all_files = glob.glob(os.path.join(path, "*.csv"))    

df_from_each_file = (pd.read_csv(f) for f in all_files)
dt_user   = pd.concat(df_from_each_file, ignore_index=True)
```

```python
dt_user
```

```python
dt_user = dt_user.replace(' ', np.nan)
```

```python
dt_user
```

```python
dt_user.isnull().sum()
```

```python
new_dt_origin = dt_user["Origin"].str.split(",", n = 1, expand = True)
  
# making separate first name column from new data frame
dt_user["Region"]= new_dt_origin[0]
  
# making separate last name column from new data frame
dt_user["Country"]= new_dt_origin[1]
```

```python
dt_user
```

```python
new_dt_travel_style = dt_user["Travel Style"].str.split("•", n = 1, expand = True)
```

```python
dt_user['Travel_Style'] = new_dt_travel_style[1]
```

```python
dt_user
```

```python
gender_series = dt_user.username.apply(lambda x: random.choice(['male', 'female'])) 
dt_user['gender'] = gender_series
```

```python
dt_user
```

```python
age_user = dt_user.username.apply(lambda x: random.choice(range(18, 65)))
dt_user['age'] = age_user
```

```python
dt_user
```

```python
dt_user
```

```python
dt_user_member_since = dt_user["Member Since"].str.split(" ", n = 2, expand = True)
```

```python
dt_user['Member_Since'] = dt_user_member_since[2]
```

```python
dt_user
```

```python
dt_user = dt_user.dropna(subset=['Region', 'Country','Travel_Style'])
```

```python
dt_user = dt_user.reset_index()
```

```python
dt_user
```

```python
dt_user['user_id'] = dt_user.index + 1
```

```python
dt_user
```

```python
dt_user.isnull().sum()
```

```python
df_user = pd.DataFrame(dt_user,columns = ['user_id','age','Travel_Style','gender','Region','Country','Member_Since'])
```
